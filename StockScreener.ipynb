```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.marketwatch.com/tools/screener/stock?exchange=all&skip=0&orderbyfield=&direction=&visiblecolumns=Symbol,CompanyName,Price,NetChange,ChangePercent,Volume&industry=Technology&peratiomin=5&peratiomax=25&marketcapmin=50&marketcapmax=1000&outperform50dayavg=1'
```

```python
StockScrenner = requests.get(url)
StockScrenner.status_code
```

```python
soup = BeautifulSoup(StockScrenner.text, 'lxml')
```

```python
type(soup)
```

```python
stocks = soup.find_all('tr', attrs={'class':'table__row'})
```

```python
stocks_database = []

for stock in stocks:
    name = stock.find('td', attrs={'class':'table__cell align--left j-CompanyName'})
    ticker = stock.find('td', attrs={'class':'table__cell j-Symbol'})
    price = stock.find('td', attrs={'class':'table__cell j-Price'})
    volume = stock.find('td', attrs={'class':'table__cell j-Volume'})
    high_52week = stock.find('td', attrs={'class':'table__cell j-HighPrice52Week'})
    low_52week = stock.find('td', attrs={'class':'table__cell j-LowPrice52Week'})
    per = stock.find('td', attrs={'class':'table__cell j-PeRatio'})
    marketcap = stock.find('td', attrs={'class':'table__cell j-MarketCap'})

    stocks_database.append({'Name': name, 'Ticker': ticker, 'Price': price, 'Volume': volume,'High_52Week': high_52week,
                            'Low_52weel': low_52week, 'P/E Ratio': per, 'MarketCap': marketcap})

df = pd.DataFrame(stocks_database)
```

```python
df = df[1:21]

df.reset_index

df
```
